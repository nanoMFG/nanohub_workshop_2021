```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img_gray = cv2.imread('../data/test_template.tif', 0)

print(type(img_gray))
imgplot = plt.imshow(img_gray, cmap='gray')
```

```python
t = 20
threshold = 0.25
template_posn = [500,275]
```

```python
img_rgb = cv2.cvtColor(img_gray, cv2.COLOR_GRAY2BGR)
cv2.rectangle(img_rgb, (template_posn[0],template_posn[1]), (template_posn[0]+t,template_posn[1]+t), (255,0,0), 2)
imgplot = plt.imshow(img_rgb, cmap="gray")
```

```python
template = img_gray[template_posn[0]:template_posn[0]+t,template_posn[1]:template_posn[1]+t]
print(type(img_gray))
template_plot = plt.imshow(template, cmap='gray')
```

```python
res = cv2.matchTemplate(img_gray, template, cv2.TM_SQDIFF_NORMED)

plt.imshow(res, cmap='gray')
```

```python
simple_mask = np.where(res>=threshold, 1, 0)

plt.imshow(simple_mask, cmap='gray')
```
